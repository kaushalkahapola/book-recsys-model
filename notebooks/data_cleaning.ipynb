```python
import pandas as pd
import numpy as np
import re
```

```python
books_original = pd.read_csv('data/Books.csv')
books_original.head()
```

```python
# Make a new column in books with all characters in lowercase and without special characters
books = books_original.copy()
books['Book-Title-Modified'] = books['Book-Title'].str.lower()
books['Book-Title-Modified'] = books['Book-Title'].str.replace(r'\W', '', regex=True)
books['Book-Title-Modified'] = books['Book-Title-Modified'].str.replace(r'\s+', ' ', regex=True)

print("Number of unique titles in the original dataset:", books['Book-Title'].nunique())
print("Number of unique modified titles in the dataset:", books['Book-Title-Modified'].nunique())


# books.to_csv('data/cleaned_books_modified.csv', index=False)
```

```python
print("Number of records in the original dataset:", books_original.shape[0])
```

```python
# Sort the dataset by author name
books = books.sort_values(by='Book-Author')
books.head(10)
```

```python
# Give common ISBN to books with the same title and author
books['ISBN_new'] = books.groupby(['Book-Title-Modified']).ngroup()
books.head(10)
```

```python
books.ISBN_new.nunique()
```

```python
books["Book-Title"].nunique()
```

```python
# Modify the ISBNs in the ratings dataset
ratings = pd.read_csv('data/Ratings.csv')

# ratings = ratings.merge(books[['ISBN', 'ISBN_new']], on='ISBN', how='left')
# ratings = ratings.drop(columns=['ISBN'])
# ratings = ratings.rename(columns={'ISBN_new': 'ISBN'})
ratings.head()
```

```python
# Create a mapping from ISBN to ISBN_new
isbn_mapping = books[['ISBN', 'ISBN_new']].drop_duplicates()
isbn_mapping.head()
```

```python
# Remove rows from books where ISBN is not in the ratings dataset
books = books[books['ISBN'].isin(ratings['ISBN'])]
books.head(10)
```

```python
print(books_original.shape, books.shape)
```

```python
# Remove rows from ratings where ISBN is not in the books dataset

ratings = ratings.merge(books, on='ISBN', how='right')
ratings.isna().sum()
```

```python
ratings.ISBN.nunique()
```

```python
ratings.ISBN_new.nunique()
```

```python
ratings_modified = ratings[['User-ID', 'ISBN_new', 'Book-Rating']]
ratings_modified = ratings_modified.rename(columns={'ISBN_new': 'Book-ID'})
ratings_modified.head()
```

```python
books_modified = books[['ISBN_new', 'Book-Title', 'Book-Author']]
books_modified = books_modified.rename(columns={'ISBN_new': 'Book-ID'})
books_modified = books_modified.drop_duplicates(subset='Book-ID')
books_modified.sort_values(by='Book-Author').head(10)
books_modified.shape
```

```python
books_modified.to_csv('data/cleaned_books_modified.csv', index=False)
ratings_modified.to_csv('data/cleaned_ratings_modified.csv', index=False)
```

```python
# Get the count of how many times each book has been rated

ratings_count = ratings_modified.groupby('Book-ID').size().reset_index(name='Rating-Count')
ratings_count = ratings_count.sort_values(by='Rating-Count', ascending=False)
```

```python
ratings_count.to_csv('data/ratings_count.csv', index=False)
```

```python
# Remove books with less than 5 ratings
ratings_reduced = ratings_count[ratings_count['Rating-Count'] >= 10]
ratings_reduced.shape
```

```python
ratings_reduced.head()
```

```python
books_reduced = books_modified[books_modified['Book-ID'].isin(ratings_reduced['Book-ID'])]
books_reduced.shape
```

```python
books_reduced.to_csv('data/cleaned_books_reduced.csv', index=False)
```

```python
ratings_reduced_df = ratings_modified[ratings_modified['Book-ID'].isin(ratings_reduced['Book-ID'])]
ratings_reduced_df.shape
```

```python
ratings_reduced_df.to_csv('data/cleaned_ratings_reduced.csv', index=False)
```
